## Some first pointers to get started

This notebook provides some first examples on how get to plotting

```python
from ggseg_py.ggseg_py import rda2gpd
from ggseg_py.plotting_utils import plot_surface, plot_view, plot_aseg
from pathlib import Path
```

```python
#load glasser

atlas_path = "../ggseg_py/atlases/glasser.rda"
gdf = rda2gpd(atlas_path, 'glasser')
gdf
```

```python
plot_surface(gdf, show_cbar=False)
```

```python
#now lets populate it with some data
import numpy as np

gdf['data2plot'] = np.arange(len(gdf))

plot_surface(gdf, column='data2plot', cmap='Reds', show_cbar=True)
```

```python
# we can also plot only a single slice
plot_view(gdf, 
          hemi='left',
          side='medial',
          column='data2plot', 
          cmap='Reds', 
          show_cbar=True)
```

```python
# the same goes for the famous aseg subcortical atlas and the classic

atlas_path = "../ggseg_py/atlases/aseg.rda"
gdf = rda2gpd(atlas_path, 'aseg')
plot_aseg(gdf)
```

```python
# and the classic desikan killany

atlas_path = "../ggseg_py/atlases/dk.rda"
gdf = rda2gpd(atlas_path, 'dk')
plot_surface(gdf, show_cbar=False)
```
